```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request
sauce3 = urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').read()
soup3=BeautifulSoup(sauce3,'lxml')
table = soup3.find('table', attrs={'class':'wikitable sortable'})
table_rows = table.find_all('tr')

res = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        res.append(row)


df = pd.DataFrame(res, columns=["Postcode", "Borough", "Neighbourhood"])
df[["Postcode","Borough","Neighbourhood"]]
df2=df["Borough"]!="Not assigned"
df3=df[df2]
#df3['Neighbourhood']=="Not assigned" 
df.loc[df['Neighbourhood'] == "Not assigned", 'Neighbourhood'] = df["Borough"]
#df3['Neighbourhood'] = np.where(df3['Neighbourhood']=="Not assigned" , df3['Borough'] , df3['Neighbourhood'])
df4=df3.groupby(['Postcode','Borough']).Neighbourhood.agg([('count'), ('Neighbourhood', ', '.join)]).reset_index()
#df4.drop(columns=['count'])
```
